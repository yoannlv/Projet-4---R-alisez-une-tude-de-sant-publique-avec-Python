```python
tx_sous_nut2 = (df_demographie_2017_dropna['Sous-nutrition'].sum()*100 / df_demographie_2017_dropna['Population'].sum()).round(2)
print("Solution 2 : Le taux de sous-nutrition dans le monde est de", tx_sous_nut2, "% de la population en 2017.")
```

# Le nombre théorique de personnes qui pourraient être nourries

Les besoins énergétiques journaliers sont entre 2400 et 2700 kcal pour un homme et entre 2000 et 2200 kcal pour une femme.
En prenant en compte le fait que les besoins d'un enfants sont moins importants, nous allons prendre arbitrairement la valeur de 2200 kcal pour un humain moyen.

```python
#Regroupement des données par pays
df_dispo_alimentaire_q2 = df_dispo_alimentaire[['Zone','Disponibilité alimentaire (Kcal/personne/jour)','Disponibilité alimentaire en quantité (kg/personne/an)','Disponibilité de matière grasse en quantité (g/personne/jour)','Disponibilité de protéines en quantité (g/personne/jour)']]
df_dispo_alimentaire_q2 = df_dispo_alimentaire_q2.groupby(['Zone'], as_index=False).sum()
df_dispo_alimentaire_q2
```

```python
#Jointure dataframe population et création de 2 colonnes
df_dispo_alimentaire_q2 = df_dispo_alimentaire_q2.merge(df_demographie_2017[['Zone','Population']])
df_dispo_alimentaire_q2['Disponibilité alimentaire totale Kcal/jour'] = df_dispo_alimentaire_q2['Disponibilité alimentaire (Kcal/personne/jour)']*df_dispo_alimentaire_q2['Population']
df_dispo_alimentaire_q2['Nombre de personnes nourrissables'] = df_dispo_alimentaire_q2['Disponibilité alimentaire totale Kcal/jour']/2200
df_dispo_alimentaire_q2
```

```python
#Sélection des pays dont la disponibilité alimentaire est < 2.200 kCal/personne/jour
df_dispo_alimentaire_pays_sous_nutri = df_dispo_alimentaire_q2[df_dispo_alimentaire_q2['Disponibilité alimentaire (Kcal/personne/jour)'] < 2200]
df_dispo_alimentaire_pays_sous_nutri.sort_values(by = 'Disponibilité alimentaire (Kcal/personne/jour)', ascending = True).head(10)
```

```python
#Nombre de pays en capacité de sous-nutrition
cnt_sous_nutri = len(df_dispo_alimentaire_pays_sous_nutri)
print("Le nombre de pays en situation de sous-nutrition est de", cnt_sous_nutri, "pays.")
```

```python
#Sélection des pays dont la disponibilité alimentaire est > 2.200 kCal/personne/jour
df_dispo_alimentaire_pays_sur_nutri = df_dispo_alimentaire_q2[df_dispo_alimentaire_q2['Disponibilité alimentaire (Kcal/personne/jour)'] >= 2200]
df_dispo_alimentaire_pays_sur_nutri.sort_values(by = 'Disponibilité alimentaire (Kcal/personne/jour)', ascending = False).head(10)
```

# Utilisation de la disponibilité intérieure

### la part qui est attribuée à l’alimentation animale

```python
#Création dataframe disponibilité alimentaire totale par pays
df_dispo_alimentaire_q4 = df_dispo_alimentaire[['Zone', 'Aliments pour animaux', 'Autres Utilisations', 'Disponibilité intérieure', 'Exportations - Quantité', 'Importations - Quantité', 'Nourriture', 'Pertes', 'Production', 'Semences', 'Traitement']]
df_dispo_alimentaire_q4 = df_dispo_alimentaire_q4.groupby(['Zone'], as_index=False).sum()
df_dispo_alimentaire_q4
```

```python
#Vérification variable
df_dispo_alimentaire_q4['Importations - Quantité'].sum() - df_dispo_alimentaire_q4['Exportations - Quantité'].sum()
```

```python
#Vérification variable
df_dispo_alimentaire_q4['Disponibilité intérieure'].sum() - ((df_dispo_alimentaire_q4['Production'].sum()) + df_dispo_alimentaire_q4['Importations - Quantité'].sum() - df_dispo_alimentaire_q4['Exportations - Quantité'].sum())
```

```python
#Calcul part consommation animale
total_alim_anim = df_dispo_alimentaire_q4['Aliments pour animaux'].sum(axis=0)
total_dispo = df_dispo_alimentaire_q4['Disponibilité intérieure'].sum(axis=0)
taux_alim_anim = (total_alim_anim / total_dispo.sum(axis=0)*100).round(2)
print("La part de la disponibilité alimentaire mondiale attribuée à l'alimentation animale est de", taux_alim_anim, "%.")
```

### la part qui est perdue

```python
#Calcul part perte
total_pertes = (df_dispo_alimentaire_q4['Pertes'].sum()).round(0)
total_pertes = int(total_pertes)
taux_pertes = ((total_pertes / total_dispo)*100).round(2)
print("Au total,", total_pertes, " milliers de tonnes sont perdues chaque année, soit", taux_pertes, "% de la disponibilité alimentaire totale.")
```

### la part qui est concrètement utilisée pour l'alimentation humaine

```python
#Calcul part consommation humaine
total_nourriture = (df_dispo_alimentaire_q4['Nourriture'].sum()).round(0)
total_nourriture = int(total_nourriture)
taux_nourriture = ((total_nourriture / total_dispo)*100).round(2)
print("Au total,", total_nourriture, " milliers de tonnes sont consacrées à l'alimentation humaine chaque année, soit", taux_nourriture, "% de la disponibilité alimentaire totale.")
```

# Répartition céréales humains / animaux

```python
#Création d'une liste 'céréales'
cereales = ['Blé', 'Riz', 'Orge', 'Maïs', 'Seigle', 'Avoine', 'Millet', 'Sorgho', 'Céréales, Autres']
```

```python
#Création d'un dataframe de disponibilité alimentaire concernant les céréales
df_dispo_cereales = df_dispo_alimentaire.loc[df_dispo_alimentaire["Produit"].isin(cereales)]
df_dispo_cereales
```

```python
#Calculs utilisation céréales
total_cereales_animaux = df_dispo_cereales['Aliments pour animaux'].sum()
total_cereales_humains = df_dispo_cereales['Nourriture'].sum()
total_cereales = df_dispo_cereales['Disponibilité intérieure'].sum()
part_cereales_animaux = ((total_cereales_animaux / total_cereales)*100).round(2)
part_cereales_humains = ((total_cereales_humains / total_cereales)*100).round(2)
print("La part de céréales consacrée aux animaux est de", part_cereales_animaux, "% tandis que celle consacrée aux humains est de", part_cereales_humains, "%.")
```

```python
print("Proportion d'alimentation animale :", "{:.2f}".format(df_dispo_cereales['Aliments pour animaux'].sum()*100/df_dispo_cereales['Disponibilité intérieure'].sum()), "%")
```

# Production et utilisation du manioc en Thaïlande

```python
#Extraction des données concernant le manioc en Thailande
thailande = df_dispo_alimentaire.loc[df_dispo_alimentaire["Zone"] == 'Thaïlande']
manioc_thailande = thailande.loc[df_dispo_alimentaire["Produit"] == 'Manioc']
manioc_thailande
```

```python
#Taux de sous-nutrition en Thailande
tx_sous_nut_thailande = df_demographie_2017.loc[df_demographie_2017['Zone'] == 'Thaïlande']
tx_sous_nut_thailande
```

```python
#Calcul du potentiel alimentaire du manioc
potentiel_manioc = int((manioc_thailande['Disponibilité alimentaire (Kcal/personne/jour)'].sum() * manioc_thailande['Production'].sum() / manioc_thailande['Disponibilité intérieure'].sum()))
print("Si l'intégralité de la production de manioc en Thaïlande était réservée à leur propre alimentation, chaque habitant pourrait consommer", potentiel_manioc, "Kcal de manioc par jour au lieu de", int(manioc_thailande['Disponibilité alimentaire (Kcal/personne/jour)'].sum()),"Kcal.")
```

```python
#Calcul du pourentage de la production du manioc sur la production totalee
part_prod_manioc_thailande = (manioc_thailande['Production'].sum() / thailande['Production'].sum()*100).round(2)
print("La part de production de manioc dans la production totale de la Thailande est de", part_prod_manioc_thailande, "%.")
```
